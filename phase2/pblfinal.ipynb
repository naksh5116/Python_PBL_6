```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("abc.csv")

# Basic inspection
print("Shape of dataset:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Handle missing values
for col in df.columns:
    if col != "City" and col != "Suitability":
        df[col].fillna(df[col].median(), inplace=True)
# Fill Suitability missing values with mode
df["Suitability"].fillna(df["Suitability"].mode()[0], inplace=True)

print("\nMissing values after fixing Suitability:")
print(df.isnull().sum())

# Verify again
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# Fix outliers using IQR method
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("Outliers handled successfully")
```

```python
# Fix data formats

# Household size as natural number
df["Household_Size"] = df["Household_Size"].round().astype(int)

# Round decimals
df["Seasonal_Index"] = df["Seasonal_Index"].round(3)
df["Garden_Area"] = df["Garden_Area"].round(3)
df["Previous_Day_Usage"] = df["Previous_Day_Usage"].round(3)
df["Daily_Water_Used"] = df["Daily_Water_Used"].round(3)

# Income as multiples of 100
df["Household_Income"] = (df["Household_Income"] / 100).round() * 100

print("Formatting done")
```

```python
# Display dataset after preprocessing
print("\nDataset after preprocessing:")
print(df.head())

print("\nDataset info:")
print(df.info())

print("\nStatistical summary:")
print(df.describe())
```
